### Lagged correlations between spike counts

```python
acorr = np.correlate(spikes-np.mean(spikes),spikes-np.mean(spikes),"full")/(N*np.std(spikes)**2)
```

```python
fig, ax = plt.subplots(2,1)
fig.set_size_inches(12,5)
ax[0].plot(delta*np.linspace(-N,N,2*N-1),acorr,'.',color = 'k')
ax[0].axis([-N,N,-.2,1.1])
ax[1].plot(delta*np.linspace(0,256,256),acorr[N-1:N+255],'.',color = 'k')
ax[1].axis([0,delta*256,-.2,1.1])
ax[1].set_xlabel('delay (ms)')
```

## Relationship between spikes and stimulus

We will consider a very simplified version of the ``decoding problem'', wherein we try to associate spiking events with events in the stimulus space by looking at the spike triggered average...

```python
def sta(frames,spikes,lag):
    laggedspikes = spikes[lag:-1]
    weighted_frames = [float(s)*frame for s,frame in zip(laggedspikes,frames)]
    return sum(weighted_frames)/len(laggedspikes)
```

```python
allframes = [np.array(stim[:,:,t]) for t in range(N)]
lags = range(16)
lagged_stas = [sta(allframes,spikes,lag) for lag in lags]
```

```python
plt.imshow(lagged_stas[0],cmap = 'gray',interpolation = 'nearest',aspect="equal")
plt.colorbar()
```

So, we see a bright spot here. Looks cool. Lets look at this as a function of the lag.

```python
fig, ax = plt.subplots(4,4)
fig.set_size_inches(10,10)
timesteps = range(3)
axs = [[plt.subplot2grid((4,4),(j,i),colspan=1,rowspan=1) for i in range(0,4)] for j in range(0,4)]
for lag,axframe in enumerate(zip(sum(axs,[]),lagged_stas)):
    ax,frame = axframe
    ax.imshow(frame,cmap = 'gray',vmin=-.2,vmax = .5,interpolation = 'nearest',aspect="equal")
    ax.set_title(lag)
    ax.axis('off')
```

There seems to be a charateristic hotspot in the center. Lets look at it as a function of time

```python
plt.plot(delta*np.array(lags),[lagged_stas[lag][7,8] for lag in lags],color='k',lw=3)
plt.xlabel('lag (ms)')
```
